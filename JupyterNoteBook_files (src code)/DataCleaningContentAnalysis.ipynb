```python
import numpy as np
import networkx as nx
import pandas as pd
import csv
```

```python
fileName = 'sortedByYearPostsData.csv.csv'
# read csv file into a pandas dataframe
df = pd.read_csv(fileName, index_col=False)

# Filter the dataframe based on year
p_df = df[df['year'] > 2019]
p_df.head(3)
```

```python
f1 = ['p_C0.csv', 'p_C1.csv', 'p_C2.csv']

dfs = []
for f in f1:
    with open(f1[0], newline='', encoding = "UTF-8") as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='"')
        for row in reader:
            dfs.append(row)
    
            
com_df = pd.DataFrame(dfs)

# Assign first row as header
com_df.columns = com_df.iloc[0]
com_df = com_df[1:] 

# com_rply_df.to_csv('content_comb.csv', index=False)
```

```python
# Merge the columns using a lambda function
com_rply_df['content'] = com_rply_df.iloc[:, 3:102].apply(lambda row: '_'.join([str(val) for val in row if pd.notnull(val)]), axis=1)
# Drop the original columns

com_rply_df.drop(com_rply_df.columns[3:102], axis=1, inplace=True)

# create a new column "merged_id" that combines "comment_id" and "reply_id" based on conditions
com_rply_df['comment_id'] = np.where(com_rply_df['comment_id'] == '', com_rply_df['reply_id'], 
                                 np.where(com_rply_df['reply_id'] == '', com_rply_df['comment_id'], 
                                          com_rply_df['comment_id'] + com_rply_df['reply_id']))

display(com_rply_df)
```
