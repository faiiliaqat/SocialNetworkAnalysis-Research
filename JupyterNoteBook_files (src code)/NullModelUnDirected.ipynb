```python
import networkx as nx
import random
import numpy as np
from datetime import datetime
import pandas as pd
#import matplotlib.pyplot as plt
from networkx import double_edge_swap
```

```python
# read csv file into a pandas dataframe
df = pd.read_csv('UnDir_edges_Comm.csv')

# create MultiDiGraph from dataframe of SubGraph
G = nx.from_pandas_edgelist(df, edge_attr=True, create_using=nx.Graph())

print(f"Created a graph with {len(G.nodes)} nodes and {len(G.edges)} links.")
```

```python
# gephi C = Average Clustering Coefficient: ?????

C = np.mean(list(nx.clustering(G).values()))

#  dW = getLargeWCC(G)
d = nx.average_shortest_path_length(G, weight='weight')

print("Average Clustering Coefficient: ", C)
print("Average Shortest Path Length: ", d)
```

```python
DP = G.copy() # the function edits the graph in place


# Let's create the ensemble.

clustering_DP = []
short_path_DP = []

count0 = 0
for i in range(1000): # 1000 is better
    print(f"[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] Starting...")
    
    # no garuntee of connectivity in this model
    nx.connected_double_edge_swap(DP,nswap=10*G.number_of_edges(),_window_threshold=3)
    C_DP = np.mean(list(nx.clustering(DP).values()))
    d_DP = nx.average_shortest_path_length(DP,weight=None)
    count0 +=1  
    short_path_DP.append(d_DP)
    clustering_DP.append(C_DP)
    
    
```

```python
print("Clustering mean:",np.mean(clustering_DP))
print("Clustering std:", np.std(clustering_DP))

print("SPL mean:", np.mean(short_path_DP))
print("SPL std:", np.std(short_path_DP))

print(count0)
```
