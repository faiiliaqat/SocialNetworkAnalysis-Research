```python
from nltk.stem import PorterStemmer


def preprocess_text(text):
    
    # check if text is NaN
    if type(text) != str:
        return ""
    

    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = text.lower() # convert text to lower-case
     # Remove non-alphanumeric characters
    text = word_tokenize(text) # remove repeated characters (helloooooooo into hello)

    stop_words = set(stopwords.words('english') + list(punctuation) + ['would', 'could', 'im' , 'got', 'get' , 'ive' , 'go'])

    # Remove stopwords and stem words
    words = [word for word in text if word not in stop_words]
    print(words)
    return ' '.join(words)
```

Tt is simply counting the frequency of each word and returning the 50 most common words in the vocabulary. This code is useful for quickly identifying the most frequent words in the corpus, while the TfidfVectorizer code is more useful for identifying the most informative words in the corpus that are specific to each document.

The nltk.FreqDist() approach counts the frequency of each word in the text and returns a frequency distribution of words. This is useful for getting an idea of what words occur most frequently in the text corpus.

```python
# Build vocabulary
def FreqWords(preprocessedTrainingData, n):
    all_words = []
    for words in preprocessedTrainingData:
        all_words.extend(words)
    wordlist = nltk.FreqDist(all_words)
    top_words = wordlist.most_common(n)
    
    for word in top_words:
        print(word)
```

If you want to find words that are important in each document and are also important across all documents in the corpus, you can use the TfidfVectorizer with default parameters, which computes the TF-IDF scores for each word in each document and across the entire corpus.

You can then use the resulting matrix to compute the average importance score for each word across all documents. Words with high average scores are important in each document and across the corpus.


The TfidfVectorizer() approach calculates the importance of each word in the text corpus based on its frequency in the individual documents and the overall corpus. This approach is useful for finding the most important words in the corpus and is often used in text classification and clustering tasks.


the TfidfVectorizer object is created and fitted to the corpus of text. The resulting matrix tfidf_matrix has shape (n_documents, n_words), where n_documents is the number of documents in the corpus and n_words is the number of unique words in the corpus.

The code then computes the average TF-IDF score for each word across all documents using the mean() method of the matrix. The resulting array avg_scores has shape (1, n_words).

The feature names (i.e., unique words) are obtained using the get_feature_names() method of the TfidfVectorizer object. The code then creates a dictionary word_scores that maps each word to its average score.

Finally, the words are sorted by average score in descending order using the sorted() function and the key argument to specify the score. The top 10 words by average score are printed using a for loop.

```python
def tfidf_vec(processedData):
    
    # Create TfidfVectorizer object
    tfidf_vectorizer = TfidfVectorizer()

    # Fit and transform the preprocessed text data
    tfidf_matrix = tfidf_vectorizer.fit_transform(processedData)

    # compute the average TF-IDF score for each word across all documents
    avg_scores = tfidf_matrix.mean(axis=0)
    
    # Get the vocabulary (i.e., unique words)
    feature_names  = tfidf_vectorizer.get_feature_names()
    
    
    # create a dictionary mapping words to average scores
    word_scores = {}
    for feature, score in zip(feature_names, avg_scores.tolist()[0]):
        word_scores[feature] = score
    
    
    # sort the words by average score (descending order)
    sorted_words = sorted(word_scores, key=word_scores.get, reverse=True)
    
    # print the top 10 words by average score
    top_words = sorted_words[:20]
    for word in top_words:
        print(word, "\t:",word_scores[word])
```
